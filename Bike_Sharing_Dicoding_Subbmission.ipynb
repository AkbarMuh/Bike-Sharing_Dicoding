### Cleaning Data

```python
# Langkah 3: Cleaning Data
# Tidak ada data cleaning yang diperlukan dalam contoh ini karena dataset bersih.

# Konversi kolom 'dteday' menjadi tipe data datetime
df['dteday'] = pd.to_datetime(df['dteday'])
```

## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Langkah 4: Exploratory Data Analysis (EDA)
# Distribusi Data
plt.figure(figsize=(15, 10))
sns.set(style="whitegrid")
sns.boxplot(data=df[['temp', 'atemp', 'hum', 'windspeed']])
plt.title('Distribusi Data Variabel Numerik')
plt.xlabel('Variabel')
plt.ylabel('Nilai')
plt.show()
```

```python
# Korelasi antar fitur numerik
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Korelasi antar Fitur Numerik')
plt.show()
```

```python
# Distribusi data untuk setiap kolom kategorikal
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='season', palette='coolwarm')
plt.title('Distribusi Data untuk Kolom Kategorikal (Musim)')
plt.xlabel('Musim')
plt.ylabel('Jumlah Data')
plt.show()
```

```python
# Melihat tren penggunaan sepeda seiring waktu
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='dteday', y='cnt')
plt.title('Tren Penggunaan Sepeda Seiring Waktu')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Peminjaman Sepeda')
plt.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Pola musiman dalam penggunaan sepeda (count) berdasarkan jam atau musim

```python
# Menghitung jumlah rata-rata penggunaan sepeda (count) per jam dalam sehari
hourly_count = df.groupby('hr')['cnt'].mean()
hourly_count
```

```python
# Visualisasi pertanyaan bisnis 1
plt.figure(figsize=(10, 6))
hourly_count.plot(kind='line', marker='o', color='skyblue')
plt.title('Average Bike Usage per Hour')
plt.xlabel('Hour')
plt.ylabel('Average Count')
plt.xticks(range(24))
plt.grid(True)
plt.show()
```

```python
print("Hasil Analisis Pertanyaan Bisnis 1:")
print("---------------------------------------------------")
print("Pola musiman dalam penggunaan sepeda (count) berdasarkan jam:")
for hour, count in hourly_count.items():
    print(f"Jam {hour}: Rata-rata penggunaan sepeda = {count:.2f}")
print("---------------------------------------------------")
```

### Pertanyaan 2: Hubungan antara temperatur dan kepadatan penggunaan sepeda (count) setiap jam

```python
# Visualisasi scatter plot antara kolom 'temp' dan 'cnt'
plt.figure(figsize=(10, 6))
plt.scatter(df['temp'], df['cnt'], alpha=0.5)
plt.title('Temperature vs Bike Usage')
plt.xlabel('Temperature')
plt.ylabel('Count')
plt.show()
```

```python
correlation = df['temp'].corr(df['cnt'])

print("\nHasil Analisis Pertanyaan Bisnis 2:")
print("---------------------------------------------------")
print("Hubungan antara temperatur dan kepadatan penggunaan sepeda (count) setiap jam:")
print(f"Korelasi antara temperatur dan penggunaan sepeda: {correlation:.2f}")
if correlation > 0:
    print("Terdapat korelasi positif antara temperatur dan penggunaan sepeda.")
elif correlation < 0:
    print("Terdapat korelasi negatif antara temperatur dan penggunaan sepeda.")
else:
    print("Tidak ada korelasi antara temperatur dan penggunaan sepeda.")
print("---------------------------------------------------")
```
